```python

# Importing the libraries

import nltk
#nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
#nltk.download('wordnet')

import re
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
from sklearn import datasets, linear_model
from sklearn.model_selection import ShuffleSplit # or StratifiedShuffleSplit
def text_preprocess(ds: pd.Series) -> pd.Series:
    """
    Apply NLP Preprocessing Techniques to the reviews.
    """
    for m in range(len(ds)):
        
        main_words = re.sub('[^a-zA-Z]', ' ', ds[m])                                      # Retain only alphabets
        main_words = (main_words.lower()).split()
        main_words = [w for w in main_words if not w in set(stopwords.words('english'))]  # Remove stopwords
        
        lem = WordNetLemmatizer()
        main_words = [lem.lemmatize(w) for w in main_words if len(w) > 1]                 # Group different forms of the same word
        
        main_words= main_words[-15:]
        
        main_words = ' '.join(main_words)
        ds[m] = main_words
    return ds
dataset = pd.DataFrame(new_data)
dataset[0] = text_preprocess(dataset[0])
# for i in dataset[0]:
#     print(i)
```

```python

def categorizer(items,category):
    cats = {}
    for item in items:
        if item not in category:
            continue
        cat = category[item]
        if cat!="other":
            if cat not in cats:
                cats[cat] = 0
            cats[cat] +=1
    maximum = -1
    val = "other"
    for cat in cats:
        if cats[cat] > maximum:
            val = cat
            maximum = cats[cat]
    #print(val)
    return val
#consider this to be a blackbox classifier that classifies the inputs to their associated labels and returns it
def classify(dataset,vectorizer,classifier):
    dataset[0] = text_preprocess(dataset[0])    
    X = dataset.iloc[:, 0].values
    X = td.transform(X)
    y_pred = classifier.predict(X)
    return y_pred

def select_and_classify((data,u_data, labels, unique, category, td, classifier, td_sel, classifier_sel):
    #data: the inputs to be classified as parsed by getall from the source html 
    #u_data: The user dictionary that contains the user data to fill 
    #labels: The labels that we support 
    #unique
    #category
    #td: vectorizer for model one(text input type classification). Vectorizer prepares input for the AI classifier
    #classifier: Classifier 1(text input typeclassification) that takes the vectorized input and produce 
    #associated labels for the text input type 
    #td_sel: vectorizer for model 2(select and radio type classification). Vectorizer prepares input for the AI classifier
    #clasifier_sel: Classifier 2(sel and radio type classification) that takes the vectorized input and produce 
    #associated labels for the select and radio type inputs. 
    
    new_data = []
    labels = labels
    sel_data = []
    idx = []
    idx_sel = []
    for i in range(len(data)):
        #filter out the text type fields from data and put them into new_data
        if data[i][4] == 'input' and data[i][5] == 'text' and data[i][-2] in ['text','email','url','date','month','tel','week']:
            idx.append(i)
            new_data.append([data[i][0],''])
        # filter out the select and radio type fields and put them into sel_data
        if (data[i][4] == 'input' and data[i][5] in ['radio','checkbox']) or (data[i][4] == "select"):
            idx_sel.append(i)
            sel_data.append([data[i][0],data[i][1],data[i][4],data[i][5]]) 
    #print(new_data) 
    print(len(data))
    print(len(idx))

    ### text input classification. consider black box
    #create pandas dataframe
    dataset = pd.DataFrame(new_data)
    #preprocess the input sentences 
    dataset[0] = text_preprocess(dataset[0]) 
    X = dataset.iloc[:, 0].values
    #vectorize the inputs
    X = td.transform(X)
    #classify the inputs
    y_pred = classifier.predict(X)
    #map the classified classes from numbers to their associated labels.
    inp_cats = [labels[i] for i in y_pred]
    out = pd.DataFrame(dataset[0])
    out[1] = y_pred


    ##select,radio,checkbox classification.
    sel_data_preproc = []
    #concatenate all the select lists into one so each element can be classified separately.
    #example: [[man,woman,lgbt],[not a veteran, yes veteran]] -> [man,woman,lgbt,not a veteran, yes veteran]
    for i in sel_data:
        sel_data_preproc.extend(i[1])
    #same as before preprocess, vectorize and classify but this time with second model for select, radio, checkbox data
    dataset = pd.DataFrame(sel_data_preproc)
    dataset[0] = text_preprocess(dataset[0])    
    X = dataset.iloc[:, 0].values
    X = td_sel.transform(X)
    y_pred_sel = classifier_sel.predict(X)

    start = 0
    inp_cats_sel = []
    for i in sel_data:
        #seperate the output from the classifier back as it was in sel_data
        #[1,0,3,4,5] -> [[1,0,3],[4,5]]
        i[1] = y_pred_sel[start:start+len(i[1])]
        #convert the numbers back into their corresponding labels
        #[[1,0,3],[4,5]] -> [[male,female,lgbt],[no veteran, yes veteran]]
        #Note that above labels are what we are trying to identify. The whole purpouse is to map man -> male, 
        #I am not a veteran -> no veteran and so on
        # unique is the labels equivalent for radio and select data
        i[1] = [unique[j] for j in i[1]]
        #identify the cateogry for each sub list
        #[man,woman,lgbt] -> gender. Category dictionary contains the mapping. The categorizer function extracts the 
        #correct mapping.
        cat = categorizer(i[1],category)
        if cat == 'month':
            cat = labels[classify(pd.DataFrame([i[0]]), td, classifier)[0]]
        #inp_cats_sel is the eqiuivalent of inp_cats.It contains the category select, radio boxes etc belong to
        inp_cats_sel.append(cat)
        start+=len(i[1])

    #preparing input(inps) and category list(tags) for postprocess and value extraction
    #the inp list is filled up from the inp_cat or inp_cat_sel or "" according to idx and idx_sel respectively
    inps=[]
    tags=[]
    for i in range(len(data)):
        if i in idx:
            inps.append(inp_cats[idx.index(i)])
            tags.append(['text',""])
        elif i in idx_sel:
    #         print(i,inp_cats_sel)
            inps.append(inp_cats_sel[idx_sel.index(i)])
    #         print(sel_data[idx_sel.index(i)])
            if sel_data[idx_sel.index(i)][2] == 'select':
                tags.append(['select',sel_data[idx_sel.index(i)][1]])
            elif sel_data[idx_sel.index(i)][3] in ['radio','checkbox']:
                tags.append(['check',sel_data[idx_sel.index(i)][1]])
        else:
            inps.append("")
            tags.append(["",""])
    #print(inps,tags)
    return postprocess(inps,u_data,tags)
```

```python
# terms:
# input field: one of the many input boxes, check boxes, or radio buttons to be filled like "first name"
# input value: The actual value for the input field ie "John" for firstname
# user data: The dictionary containing input values for the specific user.
def convert(data):
    #convert preprocess the user data to make it easier to find the input value for each identified input field.
    
    #CONV DICT:
    # the keys in the conv dict correspond to the label strings identified(classified) by the ML model.
    # the values of the conv dict show how to access the input value corresponding to classified label from the user
    # data. (Take a look at the structure of sample user data few blocks below to see its structure)
    #for example the first key value pair of the conv dict 'location or Address pers': ['address', 'personal'] tells
    # you that the value for "location or Address pers" input box is contained in the "personal" section of user_data
    # under the name "address"  (Take a look at the structure of sample user data few blocks below to see its structure)
    conv_dict = {'location or Address pers': ['address', 'personal'],
     'Full Name': ['full-name', 'personal'],
     'Age Group': ['age-group', 'personal'],
     'City': ['city', 'personal'],
     'Country': ['country', 'personal'],
     'company name current': ['current-company', 'personal'],
     'Job Title current': ['current-position', 'personal'],
     'Do you have any disabilities': ['disability-status', 'personal'],
     'Email address': ['email', 'personal'],
     'Ethnicity': ['ethnicity', 'personal'],
     'First Name': ['first-name', 'personal'],
     'Gender': ['gender', 'personal'],
     'Github url': ['github', 'personal'],
     'Last Name': ['last-name', 'personal'],
     'Are you legally authorized to work in the United States?': ['legal-status',
      'personal'],
     'Linkedin url': ['linkedin', 'personal'],
     'Phone number': ['phone', 'personal'],
     'Prefer Pronouns': ['pronouns', 'personal'],
     'State': ['state', 'personal'],
     'Are you a protected veteran': ['veteran-status', 'personal'],
     'Will you now or in the future require sponsorship for employment visa status (e.g. H-1B)': ['working-visa',
      'personal'],
     'Zip Code': ['zip-code', 'personal'],
     'Field of Study': ['field', 'education'],
     'starting From or start date educ': ['from', 'education'],
     'GPA': ['gpa', 'education'],
     'skills': ['related-skills', 'education'],
     'courses': ['relevant-courses', 'education'],
     'degree': ['state', 'education'],
     'graduation educ': ['to', 'education'],
     'School or University': ['university', 'education'],
     'company name': ['company', 'experience'],
     'starting From or start date exps': ['from', 'experience'],
     'Job Title': ['job-title', 'experience'],
     'location or Address exps': ['location', 'experience'],
     'job description': ['role-description', 'experience'],
     'end date exps': ['to', 'experience'],
     'starting From or start date exps month':['start-month','experience'],
     'starting From or start date exps year':['start-year','experience'],
     'end date exps month':['end-month','experience'],
     'end date exps year':['end-year','experience'],
     'starting From or start date educ month':['start-month','education'],
     'starting From or start date educ year':['start-year','education'],
     'graduation month':['end-month','education'],
     'graduation year':['end-year','education'],
      'other' : ['other', 'personal']}
    
    # a third element (a counter) is added to each value item of conv dict. This counter will then be incremneted each
    # time a value is accessed. For example once a company name corresponding to the first experience is accessed, 
    # The counter will help extract the company name corresponding to the second experience the next time it is accessed.
    for key,val in conv_dict.items():
    #     convert_dict[key] = [val, 'personal', 0]
        val.append(0)
    
    # creates two additional date fields for each experience and education 
    # with just the year part or just the month part. Some input boxes ask for them separately
    # similarly a full name field is added as sometimes first and last names are not asked seperately.
    exps = data['experience']
    for exp in exps:
        if exps[exp]['from']:
            exps[exp]['start-month'] = exps[exp]['from'][5:]
            exps[exp]['start-year'] = exps[exp]['from'][:4]
        if exps[exp]['to']:
            exps[exp]['end-month'] = exps[exp]['to'][5:]
            exps[exp]['end-year'] = exps[exp]['to'][:4]
    educ  = data['education']
    if educ['from']:
        educ['start-month'] = educ['from'][5:]
        educ['start-year'] = educ['from'][:4]
    if educ['to']:
        educ['end-month'] = educ['to'][5:]
        educ['end-year'] = educ['to'][:4]
    data['personal']['full-name'] = data['personal']['first-name'] + ' ' + data['personal']['last-name']
    return conv_dict, data

def get_inp(conv_dict,data, inp):
    # accesses the category, type, num from conv dict
    cat, typ, num = conv_dict[inp][0], conv_dict[inp][1], conv_dict[inp][2]
    #increases the counter for accessing the next experience or education next time
    conv_dict[inp][2] += 1
    #accesses the correct input value from the correct section from the user_data
    if typ == 'personal' or typ =='education':
        if cat not in data[typ]:
            return ""
        return data[typ][cat]
    elif typ == 'experience':
        if num not in data[typ] or cat not in data[typ][num]:
            return ""
        return data[typ][num][cat]

def postprocess(inps, u_data, tags):
    #Postprocess does the additional identification that AI model doesnt do currently. it then uses get_inp and 
    #categorizer to get the correct value 
    # Needed additional classification includes the following
    # a) furhter identify if the detected start date and end date is asking for start or end date of an enxperience or
    #    an education field 
    # b) Further Detect whehter the "location or address" is asking for "home address" or the "address of a particular
    #    the work experience"
    # c) if an "year" field is detected instead of whole date figure out what year it is asking for
    # d) detect whether the company name and job title is asking for the current job title and current company name 
    #    or the company name and title of a previous work experience 
    conv_dict, u_data = convert(u_data)
    for i in range(len(inps)):
        prev_exps = ['Job Title','company name']
        prev_ed = ['']
        
        # detect if the start date corresponds to experience or education by looking at nature of the neighbourhood
        # for example if company or job title was detected earlier then the date is probably for an experience
        if inps[i] in ['starting From or start date']:
            if inps[i-1] in prev_exps or inps[i-2] in prev_exps or inps[i-3] in prev_exps:
                inps[i] = inps[i] + ' ' + 'exps'
            elif inps[i-1] in prev_ed or inps[i-2] in prev_ed or inps[i-3] in prev_ed:
                inps[i] = inps[i] + ' ' + 'educ'
        elif inps[i] in ['end date']:
            if inps[i-2] in prev_exps or inps[i-3] in prev_exps or inps[i-4] in prev_exps:
                inps[i] = inps[i] + ' ' + 'exps' 
            elif inps[i-2] in prev_exps or inps[i-3] in prev_exps or inps[i-4] in prev_exps:
                # end date is changed to graduation
                #inps[i] = 'graduation' + ' ' + 'educ'
                inps[i] = 'graduation'
        # same algorithm as above is used for address problem b) 
        elif inps[i] in ['location or Address']:
            if inps[i-2] in prev_exps or inps[i-3] in prev_exps or inps[i-4] in prev_exps:
                inps[i] = inp[i] + ' ' + 'exps'
            else:
                inps[i] = inp[i] + ' ' + 'pers'
        # if year is detected it means that they want the month and year to be filled in different boxes so label them
        # accordingly.
        elif inps[i] in ['year']:
            if inps[i-1] in ['starting From or start date exps']:
                inps[i-1] = inps[i-1] + ' ' + 'month'
                inps[i] = 'starting From or start date exps' + ' '+ 'year'
            elif inps[i-1] in ['end date exps']:
                inps[i-1] = inps[i-1] + ' ' + 'month'
                inps[i] = 'end date exps'+ ' '+ 'year'
            elif inps[i-1] in ['starting From or start date educ']:
                inps[i-1] = inps[i-1] + ' ' + 'month'
                inps[i] = 'starting From or start date educ' + ' '+ 'year'
            elif inps[i-1] in ['graduation']:
                inps[i-1] = inps[i-1] + ' ' + 'month'
                inps[i] = 'graduation' + ' '+ 'year'
                
            
    num_title = inps.count('Job Title')
    num_name = inps.count('company name')
    num_start = inps.count('starting From or start date exps')
    num_loc = inps.count('end date exps')
    num_exps = num_start if num_start > 0 else num_loc
    
    #needs modification
    # currently checks if number of job titles detected is the same as the number of start dates 
    # if not then the first job title and company corresponds to the current company and current title
    #(dates are almost never asked for current posts so this a very hacky way to detect that)
    if num_title > num_exps:
        inps[inps.index('Job Title')] = 'Job Title current'
    if num_name > num_exps:
        inps[inps.index('company name')] = 'company name current'
    
    print(inps)
    # create a result array (size of inputs array) containing the values to be filled in.
    for i in range(len(inps)):
        result[i] = ''
        # if our model is uncertain of the classification don'tfill in anything ie ''
        if inps[i] == '' or inps[i] == 'None of the above':
            continue
        user_inp = get_inp(conv_dict, u_data, inps[i])
        print(user_inp)
        # if the detected label ie man, black, veteran etc is in the user_data fill the result with 
        # the index of the corresponding value in the select options list so that option will be chosen on the frontend
        if tags[i][0] == 'select':
            if user_inp in tags[i][1]:
                result[i] = tags[i][1].index(user_inp)
        # true will check the checkbox or radio button if the value corresponding to it is present in user_data
        elif tags[i][0] == 'check':
            if user_inp in tags[i][1]:
                result[i] = "true"
        # for text fill in the value from user_data as is. It will be filled into the input textbox.
        elif tags[i][0] == 'text':
            result[i] = user_inp
    return result
```
